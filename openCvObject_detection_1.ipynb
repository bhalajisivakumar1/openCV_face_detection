```python
import cv2
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img2 = cv2.imread(r"C:\Users\acer\Downloads\download.jfif")
```

```python
img2_new =cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2_new)
```

```python
trainedDataset = cv2.CascadeClassifier(r"E:\yolov4\8f51e58ac0813cb695f3733926c77f52\haarcascade_frontalface_default.xml")
```

```python
faces = trainedDataset.detectMultiScale(img2_new)
```

```python
faces
```

```python
for x,y,w,h in faces:
    cv2.rectangle(img2,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
cv2.imshow('ayed',img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
